```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([1,2,3,4,5,6,7])
plt.show()
```

```python

sns.distplot([1,2,3,4,5,6,7], kde=False)
plt.show()
```

```python

sns.distplot([1,2,3,4,5,6,7], hist=False)
plt.show()
```

```python
sns.distplot([1,2,3,4,5,6,7], color='red',bins=2)
plt.title('Distplot Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
